```python
fundamental_data = pd.read_csv('10K_data.csv')
price_data = pd.read_csv('./fundamental_data/price_data')
```

```python
from dateutil import parser
price_data.index = price_data['date'].apply(parser.parse)
```

```python
to_drop_for_insufficient_data = [ticker for ticker in price_data if len(price_data[ticker][price_data[ticker].isnull()]) > 250]

[price_data.drop(ticker, axis=1, inplace=True) for ticker in to_drop_for_insufficient_data]
```

```python
fundamental_data.columns
```

```python
columns = fundamental_data.columns
columns = [col for col in columns]

for col in columns[6:8]:
    fundamental_data = fundamental_data.drop(col, axis=1)
    
for col in columns[-5:]:
    fundamental_data = fundamental_data.drop(col, axis=1)

# should be inplace but ...
fundamental_data = fundamental_data.drop('Wikipedia Link', axis=1)
fundamental_data = fundamental_data.drop('Diluted EPS', axis=1)
fundamental_data = fundamental_data.drop('Beta', axis=1)
```

```python
fundamental_data = fundamental_data.drop(['Enterprise Value/Revenue', 'Enterprise Value/EBITDA'], axis=1)
```

```python
by_sectors = fundamental_data.groupby('Sector').agg([np.mean, np.std]).iloc[:,4:]
```

```python
import copy
ratio_data = copy.deepcopy(fundamental_data.iloc[:,5:])
pe_data = copy.deepcopy(ratio_data.iloc[:, [1, 5]])
ratio_data.drop(['Trailing P/E', 'Forward P/E'], axis=1, inplace=True)
```

```python
columns = ratio_data.columns
```

```python
for col in columns:
    ratio_data[col] /= ratio_data['Market Cap (intraday)']
```

```python
ratio_data = pd.concat([ratio_data, pe_data], axis=1)
```

```python
ratio_data.head()
```

```python
ratio_data['Market Cap (intraday)'] = fundamental_data['Market Cap (intraday)']
```

```python
ratio_data.index = fundamental_data['Symbol']
```

```python
ratio_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_fundamental = scaler.fit_transform(ratio_data)
```

```python
pd.DataFrame(scaled_fundamental).describe()
```

```python
from sklearn.cluster import KMeans
kmeans_transformed = KMeans(n_clusters=5).fit(scaled_fundamental)
print(kmeans_transformed.labels_)
```

```python
ratio_csv = pd.read_csv('./fundamental_data/ratio_data.csv')
```

```python
ratio_csv.head()
```

```python
ratio_csv.index
```

```python
fundamental_data.index
```

```python
ratio_csv['Sector'] = fundamental_data['Sector']
```

```python
ratio_csv.head()
```

```python
sectors = list(set(fundamental_data['Sector']))

sectors_df = {}
for sector in sectors:
    this_sector_df = copy.deepcopy(ratio_csv[ratio_csv['Sector'] == sector])
    sectors_df[sector] = this_sector_df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

for sector_key in sectors_df:
    df = sectors_df[sector_key].drop(['Sector', 'Symbol'], axis=1)
    df = scaler.fit_transform(df)
    n_clusters = min(df.shape[0], 5)
    kmeans_transformed = KMeans(n_clusters=n_clusters).fit(df)
    print(kmeans_transformed.labels_)
    
```
