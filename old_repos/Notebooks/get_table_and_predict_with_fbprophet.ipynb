```python
from postgres_utils import PostgresPandas
import pandas as pd
import numpy as np
import fbprophet as pro
```

```python
from fbprophet import Prophet
```

```python
ppd = PostgresPandas()
```

```python
tickers = ['tsla', 'aapl', 'goog', 'gs', 'gm', 'shak', 'akam']
```

```python
ret = ppd.get_tables(*tickers)
```

```python
gs = ret['gs']
```

```python
gs.head()
```

```python
gs['y'] = gs['close']
gs['ds'] = gs['date']
```

```python
m = Prophet()
m.fit(gs)
```

```python
future = m.make_future_dataframe(periods=250)
future.tail()
```

```python
forecast = m.predict(future)
forecast.tail()
```

```python
del ppd
```
